```python
# 挑選id = 13(歷史建築),  id = 15(藝文館所), id = 16(戶外踏青), id = 18(公共藝術), id = 19(親子共遊), id = 24(主題商街),進行搜尋並下載json檔

import json
import pandas as pd
from pandas import json_normalize

# 導入id = 13(歷史建築)
with open('id=13歷史建築.json', 'r', encoding = 'utf-8') as f1:
    historical = json.load(f1)

print(type(historical))
print(historical)

formatted_data = json.dumps(historical, indent = 2, sort_keys = True, ensure_ascii = False)
print(formatted_data)

df = pd.json_normalize(
    historical,
    record_path = 'data'
)
df.loc[:, ['name', 'category']]

# 導入id = 15(藝文館所)
with open('id=15藝文館所.json', 'r', encoding = 'utf-8') as f2:
    museums = json.load(f2)

print(type(museums))
print(museums)

formatted_data2 = json.dumps(museums, indent = 2, sort_keys = True, ensure_ascii = False)
print(formatted_data2)

df2 = pd.json_normalize(
    museums,
    record_path = 'data'
)
df2.loc[:, ['name', 'category']]

# 導入id=16(戶外踏青)
with open('id=16戶外踏青.json', 'r', encoding = 'utf-8') as f3:
    outside = json.load(f3)

print(type(outside))
print(outside)

formatted_data3 = json.dumps(outside, indent = 2, sort_keys = True, ensure_ascii = False)
print(formatted_data3)

df3 = pd.json_normalize(
    outside,
    record_path = 'data'
)
df3.loc[:, ['name', 'category']]

# 導入id=18(公共藝術)
with open('id=18公共藝術.json', 'r', encoding = 'utf-8') as f4:
    public_arts = json.load(f4)

print(type(public_arts))
print(public_arts)

formatted_data4 = json.dumps(public_arts, indent = 2, sort_keys = True, ensure_ascii = False)
print(formatted_data4)

df4 = pd.json_normalize(
    public_arts,
    record_path = 'data'
)
df4.loc[:, ['name', 'category']]

# 導入id=19(親子共遊)
with open('id=19親子共遊.json', 'r', encoding = 'utf-8') as f5:
    family = json.load(f5)

print(type(family))
print(family)

formatted_data5 = json.dumps(family, indent = 2, sort_keys = True, ensure_ascii = False)
print(formatted_data5)

df5 = pd.json_normalize(
    family,
    record_path = 'data'
)
df5.loc[:, ['name', 'category']]

# 導入id=24(主題商街)
with open('id=24主題商街.json', 'r', encoding = 'utf-8') as f6:
    shopping = json.load(f6)

print(type(shopping))
print(shopping)

formatted_data6 = json.dumps(family, indent = 2, sort_keys = True, ensure_ascii = False)
print(formatted_data6)

df6 = pd.json_normalize(
    shopping,
    record_path = 'data'
)
df6.loc[:, ['name', 'category']]

# 將以上六種主題的dataframe合併起來
all_df = pd.concat([df, df2, df3, df4, df5, df6], ignore_index = True)
print(all_df.shape)

# 去除重複值
cleaned_df = all_df.drop_duplicates(subset = ['name'])
print(cleaned_df['category'])
print(cleaned_df.shape)

# 將不同的id分成不同欄
list_df = cleaned_df['category'].apply(pd.Series)
list_df.columns = ['cat1', 'cat2', 'cat3', 'cat4', 'cat5', 'cat6']
print(list_df)

# 將cleaned_df, list_df合併在一起
mix_df = pd.concat([cleaned_df, list_df], axis = 1)
print(mix_df)

# mix_df.to_csv('Taipei_travel_attraction.csv', index = 0, header = 1, encoding = 'utf-8')

# 查詢各id分類的數量
content = '歷史建築'
cat1 = mix_df[mix_df.apply(lambda row: row.astype(str).str.contains(content).any(), axis = 1)]
cat1.count()

content2 = '藝文館所'
cat2 = mix_df[mix_df.apply(lambda row: row.astype(str).str.contains(content2).any(), axis = 1)]
cat2.count()
```

```python
import folium
import pandas as pd
from folium.plugins import MarkerCluster


center_lat, center_lon = 25.0478, 121.5170
m = folium.Map(location = [center_lat, center_lon], zoom_start = 15, tiles = 'Cartodb Positron')

# 台北車站(圓心)特別標示
folium.CircleMarker(
    location = [center_lat, center_lon],
    radius = 30, # 大小較大
    popup = '台北車站(圓心)',
    color = 'red',
    fill = True,
    fill_color = 'red',
    fill_cpacity = 10).add_to(m)

# 加上其他景點(用MarkerCluster管理)
df = pd.read_csv('./Taipei_travel_attraction.csv', encoding = 'utf-8')

import ast
cat = pd.Series(df['cat1'])
cat.name = 'category'
print(cat)

# 使用ast.literal_eval將cat1欄轉換成dict
cat_renew = df['cat1'].apply(ast.literal_eval)
cat_renew.name = 'category'

# 將更改格式後的cat1加回原本的DataFrame
df_cat_renew = pd.concat([df, cat_renew], axis = 1)

# 將category欄的字典提取成2欄
catel = df_cat_renew.iloc[:, 35]
cat_temp = pd.DataFrame(catel)
print('===================')
print(cat_temp)

cat_temp2 = cat_temp['category'].values.tolist()
temp2 = pd.DataFrame(cat_temp2)

cat_temp['id'] = temp2['id']
cat_temp['name'] = temp2['name']

cat_temp = cat_temp.drop(['category'], axis = 1)

# 修改欄位名稱
cat_temp.rename(columns = {'id': 'category_id', 'name': 'category_name'}, inplace = True)
cat_temp

# 將完成分欄的字典欄位加回df_cat_renew
df_cat_cleaned = pd.concat([df_cat_renew, cat_temp], axis = 1)

# 刪除cat1和category欄
df_cat_cleared = df_cat_cleaned.drop(['cat1', 'category'], axis = 1)
df_cat_cleared

# 計算各景點到圓心(台北車站)的距離
source_location = {
    'name': '台北車站',
    'Lat': 25.0478,
    'Lon': 121.5170
}

target_locations = df_cat_cleared.to_dict(orient = 'records') # 將DataFrame轉成字典清單

from geopy.distance import geodesic

def get_distances(source_location, target_locations):
    results = [
        {
            'name': target_location['name'],
            'distance(km)': geodesic((source_location['Lat'], source_location['Lon']), (target_location['nlat'], target_location['elong'])).kilometers
        }
        for target_location in target_locations
    ]
    return results

results = get_distances(source_location, target_locations)
print('~~~~~~~~~~~~~~~~~~~~~~~~~')
print(results)

# 將計算結果轉成DataFrame
distance = pd.DataFrame(results)
print(distance)
print('~~~~~~~~~~~~~~~~~~~~~~~~~')

# 依據'AttractionName'欄位，合併以上2個DataFrame
final = pd.merge(df_cat_cleared, distance, on = 'name')

# 將distance(km)取到小數點後第2位
final['distance(km)'] = final['distance(km)'].round(2)
print(final)

# final.to_csv('./Taipei_travel_attraction_cleaned.csv', index = 0, header = 1, encoding = 'utf-8')

df_updated = pd.read_csv('./Taipei_travel_attraction_cleaned.csv', encoding = 'utf-8')

marker_cluster = MarkerCluster().add_to(m)

for idx, row in df_updated.iterrows():
    folium.CircleMarker(
    location = [row['nlat'], row['elong']],
    radius = 10,
    color = 'blue' if row['category_name'] == '歷史建築' else \
            'pink' if row['category_name'] == '藝文館所' else \
            'green' if row['category_name'] == '戶外踏青' else \
            'orange' if row['category_name'] == '公共藝術' else \
            'yellow' if row['category_name'] == '親子共遊' else \
            'purple' if row['category_name'] == '主題商街' else 'black',
    fill = True,
    fill_color = 'blue' if row['category_name'] == '歷史建築' else \
            'pink' if row['category_name'] == '藝文館所' else \
            'green' if row['category_name'] == '戶外踏青' else \
            'orange' if row['category_name'] == '公共藝術' else \
            'yellow' if row['category_name'] == '親子共遊' else \
            'purple' if row['category_name'] == '主題商街' else 'black',
    fill_opacity = 1.6,
    popup = "{}，景點類型：{}，距台北車站{}公里".format(row['name'], row['category_name'], row['distance(km)'])
).add_to(marker_cluster)
m
# m.save('Taipei_station_attraction.html')
```
